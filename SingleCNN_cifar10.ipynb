```
from google.colab import drive 
import sys
sys.path.append('/content/drive/MyDrive/CNN')
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import classification_report 
from singlenet import SingleNet
from tensorflow.keras.optimizers import SGD 
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt
import numpy as np

```

Load the Data and scale it into the range[0, 1]


```
print("MESSAGE CIFAR-10 data loading ........")

((trainset_X, trainset_Y), (testset_X, testset_Y)) = cifar10.load_data()
trainset_X = trainset_X.astype("float")/255.0
testset_X = testset_X.astype("float") / 255.0
l_b = LabelBinarizer()
trainset_Y = l_b.fit_transform(trainset_Y)
testset_Y = l_b.transform(testset_Y)

labelNames = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]
```

```
print("MESSAGE! .. now compiling the model..........................")
opt = SGD(lr = 0.01)
model = SingleNet.modbuild(width = 32, height = 32, depth = 3, classes = 10)
model.compile(loss = "categorical_crossentropy", optimizer = opt, metrics = ["accuracy"])
#training the network 

print("""MESSAGE!
....now training your network.................. """)
H = model.fit(trainset_X, trainset_Y, validation_data = (testset_X, testset_Y), batch_size = 32, epochs = 40, verbose =1)
```

```

print("""MESSAGE!
....now training your network.................. """)
predictions = model.predict(testset_X, batch_size = 32)
print(classification_report(testset_Y.argmax(axis = 1), predictions.argmax(axis=1), target_names = labelNames))
```
